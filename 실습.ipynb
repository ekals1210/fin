```python
#로또 당첨 번호

import requests
html = requests.get("https://dhlottery.co.kr/gameResult.do?method=byWin")
soup = bs(html.text, "html.parser")

box = soup.find("div", {"class":"nums"})
nums = box.find_all("span")
result = []
for num in nums:
    result.append(num.text)
print(result)
```

```python
nums = box.find_all("span")
print(nums) # nums에는 span 태그에 있는 데이터 값을 빈 리스트에 정리 후 출력한다.
nums[0].text # 그 이후 리스트 첫번째 중 텍스트를 출력할 수 있게 된다.
```

```python
#네이버 인기 증시 크롤링
import requests
html = requests.get("https://finance.naver.com/sise/")
soup = bs(html.text, "html.parser")

find_ul = soup.find("ul",{"id":"popularItemList"}).find_all("li")
for j,i in enumerate(find_ul):
    print(f'{j+1} {i.find("a").text}')
```

```python
#한경 뉴스 검색 제목 크롤링

import requests
name = input("검색어를 입력하세요: ")
hankyung_url = f"https://search.hankyung.com/search/news?query={name}&page=1"
hankyung_raw = requests.get(hankyung_url)
hankyung_bs = bs(hankyung_raw.text, "html.parser")
hankyung_bs.find("div", {"class":"section_cont"})
for i in hankyung_bs.find("div", {"class":"section_cont"}).find_all("li"):
    print(i.find("em", class_="tit").text)
```
