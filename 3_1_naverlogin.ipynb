```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import pyperclip
import requests
from io import BytesIO
import base64
import getpass

def crash_captcha(image_base64, question):
    # OpenAI API 키 설정
    # OpenAI API 키 설정
    api_key = "api-key-here"
    question1 = "해당 질문에 대해 단답형으로 대답해줘 예를 들어 정답이 41 이면 41 이라고 대답하면되"

    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
    }

    payload = {
        "model": "gpt-4o",
        "messages": [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": question1
                    },
                    {
                        "type": "text",
                        "text": question
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": image_base64
                        }
                    }
                ]
            }
        ],
        "max_tokens": 1000
    }

    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

    if response.status_code == 200:
        content = response.json()['choices'][0]['message']['content']
        # 응답 출력
        print("GPT-4o 응답:", content)
        return content
    else:
        print(f"Error: {response.status_code}")
        return response.json()

def login_naver(id_, pw_):
    driver = webdriver.Chrome()
    driver.get("https://www.naver.com/")
    time.sleep(3)
    driver.find_element(By.XPATH, '//*[@id="account"]/div/a').click()
    time.sleep(2)

    pyperclip.copy(id_)
    driver.find_element(By.CSS_SELECTOR, "#id").click()
    driver.find_element(By.CSS_SELECTOR, "#id").send_keys(Keys.CONTROL, "v")
    time.sleep(2)

    pyperclip.copy(pw_)
    driver.find_element(By.CSS_SELECTOR, "#pw").click()
    driver.find_element(By.CSS_SELECTOR, "#pw").send_keys(Keys.CONTROL, "v")
    time.sleep(2)

    driver.find_element(By.CSS_SELECTOR, "#log\.login").click()
    time.sleep(3)

    if "captcha" in driver.page_source:
        print("Captcha detected, solving...")
        handle_captcha(driver, id_, pw_)

    return driver

def handle_captcha(driver, id_, pw_):
    while "captcha" in driver.page_source:
        image_base64 = driver.find_element(By.ID, "captchaimg").get_attribute("src")
        question = driver.find_element(By.ID, "captcha_info").text

        captcha_solution = crash_captcha(image_base64, question)

        pyperclip.copy(captcha_solution)
        captcha_input = driver.find_element(By.CSS_SELECTOR, "#captcha")
        captcha_input.click()
        captcha_input.send_keys(Keys.CONTROL, "v")
        time.sleep(2)

        pyperclip.copy(pw_)
        driver.find_element(By.CSS_SELECTOR, "#pw").click()
        driver.find_element(By.CSS_SELECTOR, "#pw").send_keys(Keys.CONTROL, "v")
        time.sleep(2)

        driver.find_element(By.CSS_SELECTOR, "#log\.login").click()
        time.sleep(3)

    return driver

```
