```python
import hksearch
import naver_blog
import getpass
import openai
from openai import OpenAI
```

```python
def get_gpt(user_data):
    openi_api_key = "api-key-here"
    client = OpenAI(api_key = openi_api_key)
    # "gpt-4o"
    # "gpt-4"
    message_list = [
        {"role":"system", "content": """크롤링된 기사 내용을 정리하여 투자 의견을 작성해주는 assistant 해당 내용을 블로그에 개시하기 위해 제목, 내용 으로 작성해줘 markdon 문법을 사용하지 말고 다음과 같은 형태로 작성
        다음은 작성글의 예시 이다. 제목: 블로그제목 예시\n\n내용: 블로그 내용"""},
        {"role":"user", "content":user_data}
    ]
    complection = client.chat.completions.create(
        model = "gpt-3.5-turbo",
        messages = message_list
    )
    blog_title = complection.choices[0].message.content.split("\n\n")[0].split("제목: ")[-1]
    blog_contents = complection.choices[0].message.content.split("\n\n")[1].split("내용: ")[-1]
    return blog_title, blog_contents
```

```python
query = input("검색어를 입력 하세요 : ")
page = int(input("몇페이지 까지 검색하실건가요 : "))

id_ = getpass.getpass("id: ")
pw_ = getpass.getpass("pw: ")

news = hksearch.hankyung_news(query, page)
posttitle, postdes = get_gpt(news)

driver = naver_blog.login_naver(id_, pw_)
naver_blog.posting(driver, id_, posttitle, postdes)
```
